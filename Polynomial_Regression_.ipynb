```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#import dataset
data = pd.read_csv('/content/Polynomial Regression.csv')
```

```python
data.head(5)
```

```python
data.info()
```

```python
data.describe()
```

```python
x_data = data[['x']]
y_data = data[['y']]
```

```python
# Visual Representation plot
plt.plot(figsize=(10,7))
plt.scatter(x_data,y_data)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Polynomial Regression Data')

```

```python
data.corr()
```

```python
#import LinearRegression
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
# data fit by LinearRegression
model.fit(x_data,y_data)
```

```python
y_data_predict = model.predict(x_data)
```

```python
model.score(x_data , y_data)
```

```python
#  Visual Representation plot fit and predict data
plt.scatter(x_data , y_data , label ='Polynomial Regression Data' )
plt.plot(x_data, y_data_predict , color = 'red' , label = 'fit line')
plt.xlabel('X_DATA')
plt.ylabel('Y_DATA')
plt.title ('Before Polynomial Regression DATA')
plt.legend()
```

```python
# import PolynomialFeatures
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly = PolynomialFeatures(degree=6)
```

```python
x_data_poly = poly.fit_transform(x_data)
```

```python
x_data_poly
```

```python
model.fit(x_data_poly , y_data)
```

```python
model.score (x_data_poly , y_data)
```

```python
plt.scatter(x_data, y_data, color='green')
plt.plot(x_data, 2 * np.sin(x_data), color='orange')
plt.xlabel('x_data')
plt.ylabel('y_data')
plt.title('After Polynomial Regression DATA')
```
